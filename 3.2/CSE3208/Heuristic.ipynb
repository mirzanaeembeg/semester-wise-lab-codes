Eight Queens Heuristic (No of attacking pairs)

```python
def heuristic(board):
    attacking_pair = 0
    nonattacking=0
    for i in range(len(board)):
        for j in range(i+1, len(board)):
            if board[i] == board[j] or abs(board[i] - board[j]) == abs(i - j):
                attacking_pair += 1
            if board[i] != board[j] and abs(board[i] - board[j]) != abs(i - j):
                nonattacking += 1
    print(nonattacking)
    return attacking_pair
```

```python
board = [6,1,5,7,4,3,8,1]
print('Number of attacking pairs: ', heuristic(board))
```

Eight Puzzle Heuristic (No of mismatched tiles)

```python
gtp=[(1,1,1), (2,1,2), (3,1,3), (4,2,3), (5,3,3), (6,3,2), (7,3,1), (8,2,1)]
tp=[(1,1,2), (2,1,3), (3,2,1), (4,2,3), (5,3,3), (6,2,2), (7,3,2), (8,1,1)]

i=0
h=0
while i<=7:
    if (gtp[i][1] != tp[i][1]) or (gtp[i][2] != tp[i][2]):
        h=h+1
    i=i+1

print("Heuristic 1 : ",h)
```
